# k-Nearest Neighbor (kNN)

The kNN classifier consists of two stages:

- During training, the classifier takes the training data and simply remembers it
- During testing, kNN classifies every test image by comparing to all training images and transfering the labels of the k most similar training examples
- The value of k is cross-validated

We can also use other distance metrics such as L1 distance.
For pixel values $p_{ij}^{(k)}$ at location $(i,j)$ of some image $I_k$,

the mean $\mu$ across all pixels over all images is $$\mu=\frac{1}{nhw}\sum_{k=1}^n\sum_{i=1}^{h}\sum_{j=1}^{w}p_{ij}^{(k)}$$
And the pixel-wise mean $\mu_{ij}$ across all images is
$$\mu_{ij}=\frac{1}{n}\sum_{k=1}^np_{ij}^{(k)}.$$
The general standard deviation $\sigma$ and pixel-wise standard deviation $\sigma_{ij}$ is defined similarly.

Which of the following preprocessing steps will not change the performance of a Nearest Neighbor classifier that uses L1 distance? Select all that apply. To clarify, both training and test examples are preprocessed in the same way.

1. Subtracting the mean $\mu$ ($\tilde{p}_{ij}^{(k)}=p_{ij}^{(k)}-\mu$.)
2. Subtracting the per pixel mean $\mu_{ij}$  ($\tilde{p}_{ij}^{(k)}=p_{ij}^{(k)}-\mu_{ij}$.)
3. Subtracting the mean $\mu$ and dividing by the standard deviation $\sigma$.
4. Subtracting the pixel-wise mean $\mu_{ij}$ and dividing by the pixel-wise standard deviation $\sigma_{ij}$.
5. Rotating the coordinate axes of the data, which means rotating all the images by the same angle. Empty regions in the image caused by rotation are padded with a same pixel value and no interpolation is performed.

$\color{pink}{\textit Your Answer:}$

Steps number 1, 2, and 3.

$\color{pink}{\textit Your Explanation:}$

1- Subtracting the global mean, $\mu$, from all pixel values centers the data, which is a common normalization step. Since k-NN with L1 distance relies on relative differences in pixel intensities between images, shifting all pixels by the same constant does not impact the calculated distances, leaving the classifier's performance unaffected.

2- Similarly, subtracting the pixel-wise mean, $\mu_{ij}$, centers each pixel across the dataset independently. This operation does not affect L1-based k-NN classification because the relative pixel intensity differences between images remain consistent.

3- Subtracting the mean and then scaling by the standard deviation, $\sigma$, adjusts the data to have zero mean and unit variance. While this transformation normalizes the data, it effectively just rescales the pixel values uniformly. Since L1 distance simply measures the absolute differences between pixel intensities, scaling the entire dataset does not change the nearest neighbors, thus maintaining the classifier’s performance.

4- Dividing by the pixel-wise standard deviation, $\sigma_{ij}$, changes each pixel’s scale individually, potentially altering the relative importance of different pixels in the L1 distance calculations. This can affect nearest-neighbor relationships, which could change the classifier's performance.

5- Rotating all images by a fixed angle changes the pixel coordinates, leading to different spatial relationships in the images. This transformation disrupts the direct comparison of corresponding pixel intensities, which can significantly impact L1 distance calculations and thus affect classification results.

Explain how you have implemented compute_distances_no_loops

$\color{pink}{\textit Your Answer:}$

The compute_distances_no_loops function calculates Euclidean distances between each test point in X and each training point in self.X_train without using explicit loops. It does this by vectorizing the Euclidean distance formula using matrix operations. First, it computes the squared norms of all test points (test_square) and training points (train_square) separately. These squared norms are stored in vectors that represent the sum of squared values for each data point. Then, it calculates the cross_term by taking the dot product between X and the transpose of self.X_train, resulting in a matrix of pairwise products between test and training samples.

Using these values, the function applies the expanded Euclidean distance formula $( ||a - b||^2 = ||a||^2 + ||b||^2 - 2 \cdot a \cdot b )$ to calculate the squared distances and then takes the square root to obtain the final Euclidean distances. This approach creates a fully vectorized distance matrix where each entry [i, j] is the distance between test point i and training point j, eliminating the need for nested loops and making the computation more efficient.

### Cross-validation

We have implemented the k-Nearest Neighbor classifier but we set the value k = 5 arbitrarily. We will now determine the best value of this hyperparameter with cross-validation.

```python
num_folds = 5
k_choices = [1, 3, 5, 8, 10, 12, 15, 20, 50, 100]

X_train_folds = []
y_train_folds = []
################################################################################
# TODO:                                                                        #
# Split up the training data into folds. After splitting, X_train_folds and    #
# y_train_folds should each be lists of length num_folds, where                #
# y_train_folds[i] is the label vector for the points in X_train_folds[i].     #
# Hint: Look up the numpy array_split function.                                #
################################################################################
# *****START OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)*****

X_train_folds = np.array_split(X_train, num_folds)
y_train_folds = np.array_split(y_train, num_folds)

# *****END OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)*****

# A dictionary holding the accuracies for different values of k that we find
# when running cross-validation. After running cross-validation,
# k_to_accuracies[k] should be a list of length num_folds giving the different
# accuracy values that we found when using that value of k.
k_to_accuracies = {}


################################################################################
# TODO:                                                                        #
# Perform k-fold cross validation to find the best value of k. For each        #
# possible value of k, run the k-nearest-neighbor algorithm num_folds times,   #
# where in each case you use all but one of the folds as training data and the #
# last fold as a validation set. Store the accuracies for all fold and all     #
# values of k in the k_to_accuracies dictionary.                               #
################################################################################
# *****START OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)*****

for k in k_choices:
    k_to_accuracies[k] = []

    for i in range(num_folds):

        X_val_fold = X_train_folds[i]
        y_val_fold = y_train_folds[i]

        X_train_fold = np.concatenate([fold for j, fold in enumerate(X_train_folds) if j != i])
        y_train_fold = np.concatenate([fold for j, fold in enumerate(y_train_folds) if j != i])

        knn = KNearestNeighbor()
        knn.train(X_train_fold, y_train_fold)

        y_val_pred = knn.predict(X_val_fold, k=k)

        accuracy = np.mean(y_val_pred == y_val_fold)
        k_to_accuracies[k].append(accuracy)

# *****END OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)*****

# Print out the computed accuracies
for k in sorted(k_to_accuracies):
    for accuracy in k_to_accuracies[k]:
        print('k = %d, accuracy = %f' % (k, accuracy))
```

```python
# plot the raw observations
for k in k_choices:
    accuracies = k_to_accuracies[k]
    plt.scatter([k] * len(accuracies), accuracies)

# plot the trend line with error bars that correspond to standard deviation
accuracies_mean = np.array([np.mean(v) for k,v in sorted(k_to_accuracies.items())])
accuracies_std = np.array([np.std(v) for k,v in sorted(k_to_accuracies.items())])
plt.errorbar(k_choices, accuracies_mean, yerr=accuracies_std)
plt.title('Cross-validation on k')
plt.xlabel('k')
plt.ylabel('Cross-validation accuracy')
plt.show()
```

```python
# TODO: Based on the cross-validation results above, choose the best value for k,
# retrain the classifier using all the training data, and test it on the test
# data. You should be able to get above 28% accuracy on the test data.
best_k = k_choices[np.argmax(accuracies_mean)]

knn = KNearestNeighbor()
knn.train(X_train, y_train)

y_test_pred = knn.predict(X_test, k=best_k)

num_correct = np.sum(y_test_pred == y_test)
num_test = y_test.shape[0]
accuracy = num_correct / num_test

print('%d / %d were correct => accuracy: %f' % (num_correct, num_test, accuracy))
```

**Inline Question 3**

Which of the following statements about $k$-Nearest Neighbor ($k$-NN) are true in a classification setting, and for all $k$? Select all that apply.
1. The decision boundary of the k-NN classifier is linear.
2. The training error of a 1-NN will always be lower than or equal to that of 5-NN.
3. The test error of a 1-NN will always be lower than that of a 5-NN.
4. The time needed to classify a test example with the k-NN classifier grows with the size of the training set.
5. None of the above.

$\color{pink}{\textit Your Answer:}$

statements number 2 and 4 are correct.

$\color{pink}{\textit Your Explanation:}$

1- False. The k-NN classifier does not produce linear decision boundaries. If the classes are arranged in non-linear shapes, such as concentric circles, the classifier will create boundaries that follow these curved patterns rather than straight lines.

2- True. A 1-NN classifier will always have a training error that is less than or equal to that of a 5-NN classifier. This is because, in 1-NN, each training point is its own closest neighbor, resulting in zero error for that sample. Consequently, 1-NN has the minimum possible training error.

3- False. The test error for 1-NN is not necessarily lower than for 5-NN. Consider a scenario where the training labels are [1, 0, 0, 0, 0, 0]. For a test point near the sample labeled 1, 1-NN would predict 1, which could be incorrect if the true label is 0. Meanwhile, 5-NN would likely predict 0 due to the majority vote, thus yielding a lower error. This shows that the optimal value of k depends on the specific data distribution, making cross-validation important.

4- True. The time required to classify a test example in k-NN increases with the training set size because every test point has to be compared to all training points to determine the nearest neighbors. This scaling issue can be mitigated using techniques like approximate nearest neighbors (e.g., k-d trees or ball trees), which speed up the search process by reducing the number of comparisons.
